```python
import pandas as pd

dataset = pd.read_csv('zomato.csv')
```

```python
location = pd.DataFrame(dataset.groupby('location')['location'].count().sort_values(ascending=False))[:10]


location['label'] = location.index  # Location names as label
location['count'] = location['location']  # Count values as numerical


location = location[['label', 'count']]
plt.figure(figsize=(8, 8))
plt.pie(location['count'], labels=location['label'], autopct='%1.1f%%', startangle=140)
plt.title('Location Distribution of Various Hotels in City Areas')
plt.axis('equal') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already defined and contains a 'rate' column

# Example DataFrame (replace with your actual data)
data = {'rate': [1.5, 2.2, 3.1, 4.5, 1.8, 2.7, 3.8, 4.2, 2.5, 3.3]}
df = pd.DataFrame(data)

# Ensure 'rate' column is numeric
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
df = df.dropna(subset=['rate'])  # Remove NaN values after conversion

slices = [((df['rate'] >= 1) & (df['rate'] < 2)).sum(),
          ((df['rate'] >= 2) & (df['rate'] < 3)).sum(),
          ((df['rate'] >= 3) & (df['rate'] < 4)).sum(),
          (df['rate'] >= 4).sum()
          ]

labels = ['1-2', '2-3', '3-4', '>4']  # More descriptive labels
colors = ['red', 'blue', 'green', 'yellow']
plt.figure(figsize=(15, 10))  # Create a figure before the pie chart
plt.pie(slices, colors=colors, labels=labels, autopct='%1.0f%%', pctdistance=0.5, labeldistance=1.2, shadow=True)
plt.title("Percentage of Restaurants according to their ratings")
plt.show()
```
