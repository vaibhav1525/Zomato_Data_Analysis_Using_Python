```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'rate': [3.5, 4.2, 2.8, 4.8, 3.9],
        'approx_cost(for two people)': [500, 800, 300, 1200, 600]}
df = pd.DataFrame(data)


plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='rate', y='approx_cost(for two people)', hue="rate", size="rate", palette='viridis', sizes=(10, 100))
plt.ylabel("Approx Cost (for two people)", size=16)
plt.xlabel("Restaurant Rate", size=16)
plt.title("Restaurant Rate vs. Approximate Cost (for two people)") # Added a title
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already defined and contains 'rate', 'cost', and 'online_order' columns

# Example DataFrame (replace with your actual data)
data = {'rate': [3.5, 4.2, 2.8, 4.8, 3.9],
        'cost': [500, 800, 300, 1200, 600],
        'online_order': ['Yes', 'No', 'Yes', 'Yes', 'No']}
df = pd.DataFrame(data)

plt.figure(figsize=(15, 7))
sns.scatterplot(x="rate", y='cost', hue='online_order', data=df)
plt.xticks(rotation=90)
plt.xlabel("Rate")  # Added x-axis label
plt.ylabel("Cost")  # Added y-axis label
plt.title("Rate vs. Cost with Online Order Status")  # Added a title
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'votes_count': [50, 100, 20, 150, 80, 60, 120, 40, 90, 70],
        'restaurant_rating': [3.5, 4.2, 2.8, 4.8, 3.9, 3.2, 4.1, 3.0, 4.0, 3.7]}
df = pd.DataFrame(data)
df['votes_count'] = pd.to_numeric(df['votes_count'], errors='coerce')
df['restaurant_rating'] = pd.to_numeric(df['restaurant_rating'], errors='coerce')

df = df.dropna(subset=['votes_count', 'restaurant_rating'])

sns.scatterplot(data=df, x='votes_count', y='restaurant_rating', color='purple')
plt.title('Votes vs Restaurant Ratings')
plt.xlabel('Votes Count')
plt.ylabel('Restaurant Rating')
plt.show()

```
