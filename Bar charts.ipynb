```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('zomato.csv')
top_locations = dataset['location'].value_counts().nlargest(10)

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette='viridis')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.title('Top 10 Locations by Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Zomato-style dataset
data = pd.DataFrame({
    'restaurant_name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'location': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi'],
    'cuisine': ['North Indian', 'Chinese', 'Italian', 'South Indian', 'Chinese', 'Italian'],
    'cost_for_two': [500, 800, 750, 300, 650, 700],
    'online_order': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes'],
    'rating': [4.1, 3.5, 4.3, 3.0, 4.0, 3.9]
})

sns.set(style='whitegrid')

plt.figure(figsize=(6, 4))
sns.histplot(data['rating'], bins=5, kde=True)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='cuisine', y='cost_for_two', data=data)
plt.title('Average Cost for Two by Cuisine')
plt.xticks(rotation=45)
plt.ylabel('Cost for Two')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='location', data=data, palette='Set2')
plt.title('Number of Restaurants by Location')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x='online_order', y='rating', data=data)
plt.title('Rating vs Online Order Availability')
plt.show()
```
